## Implement Hybrid RAG System

### Subtask:
Develop the core hybrid Retrieval Augmented Generation (RAG) system. This involves integrating the base language model with the real-time search component. The system should take a user query, perform a search, retrieve relevant external data, and then use the base model to generate a grounded response based on both the query and the retrieved information.


**Reasoning**:
The subtask requires defining a `hybrid_rag_system` function that integrates the real-time search with the base language model. This step will implement the function as instructed, including calling the search function, constructing the prompt, making an API call to the base model, and returning the response.



````python
import os

def hybrid_rag_system(user_query: str) -> str:
  """
  Integrates real-time search with the base language model to generate a grounded response.

  Args:
    user_query (str): The user's query.

  Returns:
    str: A grounded response from the base model.
  """
  print(f"User Query: {user_query}")

  # 1. Perform real-time search
  search_results = get_realtime_search_results(user_query)
  print("\n--- Search Results ---")
  print(search_results)
  print("----------------------\n")

  # 2. Construct comprehensive prompt for the base model
  prompt_with_search_results = f"""
  Based on the following search results and your own knowledge, please answer the user's query.

  Search Results:
  {search_results}

  User Query: {user_query}
  """

  # 3. Make an API call to the base model
  try:
    # Assuming 'client' is already defined from a previous step (first subtask)
    # and 'stream=True' is set up. Need to modify for non-streaming if a single response is desired.
    # For this RAG system, we want a single, complete response.
    completion = client.chat.completions.create(
      model="nvidia/llama-3.3-nemotron-super-49b-v1.5", # Using the same model as initialized
      messages=[
        {"role": "system", "content": "You are a helpful AI assistant that provides grounded answers based on provided information."},
        {"role": "user", "content": prompt_with_search_results}
      ],
      temperature=0.6,
      top_p=0.95,
      max_tokens=1024, # Limit token to prevent overly long responses for RAG
      frequency_penalty=0,
      presence_penalty=0,
      stream=False # Request a non-streaming response for a complete answer
    )

    # 4. Extract the generated response content
    if completion.choices and completion.choices[0].message and completion.choices[0].message.content:
      model_response = completion.choices[0].message.content.strip()
      return model_response
    else:
      return "No response generated by the model."

  except Exception as e:
    print(f"Error calling base model: {e}")
    return f"Error: {e}"

# Test the hybrid_rag_system function
sample_rag_query = "What are the latest developments in AI technology?"
rag_response = hybrid_rag_system(sample_rag_query)

print("\n--- Hybrid RAG System Response ---")
print(rag_response)
print("----------------------------------")
````

## Define Evaluation Metrics

To effectively compare the hybrid RAG system with a baseline Language Model (LLM), we will establish a set of metrics covering both performance and environmental impact. Below are the defined metrics and their proposed measurement or estimation methods.

### Performance Comparison Metrics:

1.  **Answer Accuracy**:
    *   **Definition**: How often the system provides factually correct answers that align with verifiable information.
    *   **Measurement/Estimation**: For a set of predefined test queries, responses will be manually reviewed and compared against a ground truth or expert-verified answers. A binary score (correct/incorrect) or a granular score (e.g., 0-5) will be assigned.

2.  **Answer Relevance**:
    *   **Definition**: How well the generated answer addresses the user's query and provides pertinent information.
    *   **Measurement/Estimation**: Responses to the same set of test queries will be evaluated by human assessors for their relevance to the original query, possibly using a Likert scale (e.g., 1-5, from not relevant to highly relevant).

3.  **Latency**:
    *   **Definition**: The total time taken for the system to process a user query and generate a complete response.
    *   **Measurement/Estimation**: This will be quantitatively measured using Python's `time` module (e.g., `time.time()`) to record the duration from the initiation of a query to the receipt of the final response for both systems, averaged over multiple queries.

### Carbon Footprint Assessment Metrics:

1.  **Computational Resources (CPU/GPU Usage)**:
    *   **Definition**: The amount of processing power (CPU and/or GPU cycles) consumed during the operation of the system to answer queries.
    *   **Measurement/Estimation**: Will be estimated by monitoring the average CPU and GPU utilization (if applicable) and processing time for a given set of queries. Tools like `psutil` can be used for local monitoring, or cloud provider metrics (e.g., AWS CloudWatch, Google Cloud Monitoring) for cloud-based deployments.

2.  **API Call Volume**:
    *   **Definition**: The number of external API calls made by the system (e.g., to SerpAPI for search, or to the LLM provider for completions).
    *   **Measurement/Estimation**: This can be precisely measured by logging or incrementing counters for each API request made to external services within the `hybrid_rag_system` and the baseline LLM interaction functions.

3.  **Data Transferred**:
    *   **Definition**: The total volume of data (in bytes) sent and received over the network during query processing, contributing to network energy consumption.
    *   **Measurement/Estimation**: Will be estimated by summing the sizes of request and response payloads for all external API calls. For general network traffic, tools like `netstat` or network monitoring libraries could provide broader estimates, but for this comparison, focusing on API payload sizes will be more direct.

````python
import gradio as gr
import os
import functools
import time
from openai import OpenAI
from langchain_community.utilities import SerpAPIWrapper

# --- CONFIGURATION ---
# Replace these with your actual keys
NVIDIA_API_KEY = "YOUR_NVIDIA_API_KEY"
SERPAPI_API_KEY = "YOUR_SERPAPI_API_KEY"
# ---------------------

client = OpenAI(
  base_url = "https://integrate.api.nvidia.com/v1",
  api_key = NVIDIA_API_KEY
)

os.environ["SERPAPI_API_KEY"] = SERPAPI_API_KEY
search = SerpAPIWrapper()

@functools.lru_cache(maxsize=128)
def get_realtime_search_results(query: str) -> str:
  try:
    results = search.run(query)
    if isinstance(results, list) and results:
        formatted = []
        for i, res in enumerate(results):
            title = res.get('title', 'N/A')
            link = res.get('link', 'N/A')
            formatted.append(f"Result {i+1}: {title} ({link})")
        return "\n".join(formatted)
    return str(results)
  except Exception as e:
    return f"Search Error: {e}"

def hybrid_rag_system(user_query: str):
  start_time = time.time()
  context = get_realtime_search_results(user_query)
  prompt = f"Context: {context}\n\nQuestion: {user_query}"

  try:
    completion = client.chat.completions.create(
      model="nvidia/llama-3.3-nemotron-super-49b-v1.5",
      messages=[
        {"role": "system", "content": "Use the context to answer."},
        {"role": "user", "content": prompt}
      ],
      stream=False
    )
    return completion.choices[0].message.content, context, completion.usage.total_tokens, time.time() - start_time
  except Exception as e:
    return f"Error: {e}", context, 0, 0

def baseline_llm_system(user_query: str):
  start_time = time.time()
  try:
    completion = client.chat.completions.create(
      model="nvidia/llama-3.3-nemotron-super-49b-v1.5",
      messages=[{"role": "user", "content": user_query}],
      stream=False
    )
    return completion.choices[0].message.content, completion.usage.total_tokens, time.time() - start_time
  except Exception as e:
    return f"Error: {e}", 0, 0

def compare(query):
    r_ans, r_con, r_tok, r_lat = hybrid_rag_system(query)
    b_ans, b_tok, b_lat = baseline_llm_system(query)
    return (
        f"RAG (Latency: {r_lat:.2f}s, Tokens: {r_tok}):\n{r_ans}",
        r_con,
        f"Baseline (Latency: {b_lat:.2f}s, Tokens: {b_tok}):\n{b_ans}"
    )

with gr.Blocks() as demo:
    gr.Markdown("## RAG vs Baseline Comparison")
    q = gr.Textbox(label="User Query")
    btn = gr.Button("Run Comparison")
    with gr.Row():
        out_r = gr.Textbox(label="RAG Response")
        out_c = gr.Textbox(label="Context Retreived")
    out_b = gr.Textbox(label="Baseline Response")
    btn.click(compare, inputs=q, outputs=[out_r, out_c, out_b])

demo.launch(debug=True)
````
